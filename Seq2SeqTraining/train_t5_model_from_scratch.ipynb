```python
from typing import List,Tuple, Union
import numpy as np

# Define type aliases and NamedTuples
TextInput = str
PreTokenizedInput = List[str]
EncodedInput = List[int]
TextInputPair = Tuple[str, str]
PreTokenizedInputPair = Tuple[List[str], List[str]]
EncodedInputPair = Tuple[List[int], List[int]]

lang_map_id = {
    "<am>" : 15000,
    "<tig>" : 15001,
    "<geez>" : 15002,
    "<tigre>" : 15003,
}

class GT5Tokenizer(MT5Tokenizer):
    def __init__(self, **kwargs)-> None:
        super().__init__(**kwargs)


    def __call__(self, text: Union[TextInput, PreTokenizedInput, List[TextInput], List[PreTokenizedInput]] = None,
                 lang:  Union[TextInput, List[TextInput]]="<am>", **kwargs):
        if text is not None:
          if type(text) == TextInput or type(text) == str:
            if type(lang)!=str:
              raise Exception(f"lang identifier can not be {type(lang)} for text type str")
            elif lang_map_id.get(lang, 0) == 0:
              raise Exception(f"language token '{lang}' is not supprted")
            text = lang + text
          elif type(text)== list:
            if len(text)==0:
              raise Exception("empty list")
            elif type(text[0])==str:
              langs =[]
              if type(lang)==str:
                langs = [lang for i in range(len(text))]
              elif type(lang)==list and len(lang)==len(text):
                langs=lang
              else:
                raise Exception(f"lang identifier must have equal length {len(lang)} with text {len(text)}")

              count = 0
              while count < len(text):
                if lang_map_id.get(langs[count], 0) == 0:
                  raise Exception(f"language token '{langs[count]}' is not supprted")
                text[count] = langs[count] + text[count]
                count +=1
            elif type(text[0])==list and (type(lang)==str or len(lang)==len(text)):
              if type(lang)==str:
                if lang_map_id.get(lang, 0) == 0:
                    raise Exception(f"language token '{lang}' is not supprted")
                count = 0
                while count < len(text):
                  text[count].insert(0, lang)
                  count +=1
              else:
                count = 0
                while count < len(text):
                  if lang_map_id.get(lang[count], 0) == 0:
                      raise Exception(f"language token '{lang[count]}' is not supprted")
                  text[count].insert(0, lang[count])
                  count +=1
            else:
              raise Exception(f"unsupported input type text: {type(text)}, lang: {type(lang)}")
          else:
            raise Exception(f"unsupported input type {type(text)}")

        # Call the parent class's __call__ method with the modified kwargs
        return super().__call__(text, **kwargs)
```

```python
#@title This one too

def summarizeText(text, model=model):
    text_encoding = tokenizer(
        text,
        max_length=512,
        padding='max_length',
        truncation=True,
        return_attention_mask=True,
        add_special_tokens=True,
        return_tensors='pt'
    )
    generated_ids = model.generate(
        input_ids=text_encoding['input_ids'].to(device),
        attention_mask=text_encoding['attention_mask'].to(device),
        max_length=513,
        num_beams=4,
        repetition_penalty=2.5,
        length_penalty=1.0,
        early_stopping=True
    )

    preds = [
            tokenizer.decode(gen_id, skip_special_tokens=True, clean_up_tokenization_spaces=True)
            for gen_id in generated_ids
    ]
    return "".join(preds)
```

```python
import numpy as np

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
    # Replace -100 in the labels as we can't decode them.
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    wer = wer_metric.compute(predictions=decoded_preds, references=decoded_labels)
    cer = cer_metric.compute(predictions=decoded_preds, references=decoded_labels)

    result = metric.compute(predictions=decoded_preds, references=np.array(decoded_labels).reshape(-1, 1).tolist())

    return  {"wer" : wer, "cer" : cer, "bleu": result["score"]}
```

```python
import wave
import contextlib
import os


fname = '01_d501021.wav'

filesCount = os.listdir("/content/drive/MyDrive/NewModel/ALFFAAmharic/train/wav")

hours = 0
minutes = 0
seconds = 0
for a in filesCount:
        if not a.endswith(".wav"):
            continue
        try:
          with contextlib.closing(wave.open(f"/content/drive/MyDrive/NewModel/ALFFAAmharic/train/wav/{a}",'r')) as f:
                  frames = f.getnframes()
                  rate = f.getframerate()
                  duration = frames / float(rate)
                  seconds += duration
                  if seconds >= 60:
                          minutes += seconds// 60
                          seconds =  seconds%60

                  if minutes >= 60:
                          hours += minutes//60
                          minutes = minutes%60
        except:
          continue
print(f"Total Training Wav Duration:{hours} hr, {minutes} mins, and {seconds} sec")
```
