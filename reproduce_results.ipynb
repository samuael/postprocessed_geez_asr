```python
data = librosa.load(path="01_d501021.wav", sr=16000)
```

```python
from transformers import Wav2Vec2FeatureExtractor

feature_extractor = Wav2Vec2FeatureExtractor(feature_size=1, sampling_rate=16000, padding_value=0.0, do_normalize=True, return_attention_mask=False)
```

```python
# from urllib.request import urlretrieve
# urlretrieve(("https://huggingface.co/Samuael/asr-amharic-phoneme-based-37-6/blob/main/vocab.json"), "vocab.json")
```

```python
from transformers import Wav2Vec2CTCTokenizer
tokenizer = Wav2Vec2CTCTokenizer.from_pretrained("Samuael/asr-amharic-phoneme-based-37-6")
```

```python
from transformers import Wav2Vec2Processor
processor = Wav2Vec2Processor(feature_extractor=feature_extractor, tokenizer=tokenizer)
```

```python
inputs = {
    "input_values": processor(data, sampling_rate=16000),
}
```
